```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_nas = pd.read_csv("nasdaq.csv")
data_nas = data_nas.loc[:, "날짜":"종가"]
data_nas = data_nas.loc[::-1].reset_index(drop=True)
data_nas["종가"].replace('(,)', '', regex = True, inplace=True)
data_nas["종가"] = data_nas["종가"].astype(float)
data_nas["종가"] = data_nas["종가"].pct_change()
data_nas = data_nas.iloc[150:]
data_nas.columns=['DATE', '종가']
data_nas = data_nas.set_index('DATE')
data_nas
```

```python
data_gold = pd.read_csv("gold_price.csv")
data_gold = data_gold.loc[:, "날짜":"종가"]
data_gold = data_gold.loc[::-1].reset_index(drop=True)
data_gold["종가"].replace('(,)', '', regex = True, inplace=True)
data_gold["종가"] = data_gold["종가"].astype(float)
data_gold["종가"] = data_gold["종가"].pct_change()
data_gold = data_gold.iloc[1:]
data_gold.columns=['DATE', '종가']
data_gold = data_gold.set_index('DATE')
data_gold
```

```python
gold_list = list(data_gold.index)
nas_list = list(data_nas.index)
for x in range(0, 607):
    if nas_list[x] != gold_list[x]:
        data_gold = data_gold.drop(index = gold_list[x])
        del gold_list[x]
        x -= 1
data_gold
```
